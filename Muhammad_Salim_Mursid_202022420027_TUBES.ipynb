## Persiapan Dataset

### Subtask:
Memeriksa keseimbangan kelas pada kolom target, menerapkan teknik penanganan ketidakseimbangan (jika perlu), dan membagi dataset menjadi data training dan testing dengan proporsi 70:30.


```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Identify Features (X) and Target (y)
X = df.drop('Risk_Flag', axis=1)
y = df['Risk_Flag']

print("Features (X) and Target (y) separated successfully.")

# 2. Check Class Imbalance
print("\nClass distribution of the target variable (Risk_Flag):")
print(y.value_counts())
print("\nClass proportion:")
print(y.value_counts(normalize=True))
```

```python
print("Before One-Hot Encoding:")
print(X.head())

# 3. Penanganan Fitur Kategorikal (One-Hot Encoding)
categorical_cols_X = X.select_dtypes(include=['object']).columns
X = pd.get_dummies(X, columns=categorical_cols_X, drop_first=True)

print("\nAfter One-Hot Encoding:")
print(X.head())
print("\nShape of X after One-Hot Encoding:", X.shape)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 4. Bagi Dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print("Dataset successfully split into training and testing sets:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

print("\nClass distribution in y_train (after split):")
print(y_train.value_counts(normalize=True))
print("\nClass distribution in y_test (after split):")
print(y_test.value_counts(normalize=True))
```

```python
try:
    from imblearn.over_sampling import SMOTE
except ImportError:
    print("imblearn not found, installing...")
    !pip install imblearn
    from imblearn.over_sampling import SMOTE

# 5. Terapkan Oversampling (SMOTE) pada Data Pelatihan
print("Applying SMOTE to the training data...")
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("SMOTE applied successfully.")
print("\nClass distribution in y_train after SMOTE:")
print(y_train_resampled.value_counts(normalize=True))
```

## Pemodelan dan Evaluasi

### Subtask:
Memilih model klasifikasi yang sesuai (misalnya, Logistic Regression, Decision Tree, atau Random Forest), melatih model menggunakan data training, dan mengevaluasi performanya menggunakan metrik klasifikasi umum seperti akurasi, presisi, recall, F1-score, dan confusion matrix.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# 1. Inisialisasi model Logistic Regression
model_lr = LogisticRegression(random_state=42, solver='liblinear') # Using 'liblinear' solver for better convergence on small datasets

# 2. Latih model menggunakan data pelatihan yang telah di-resample
print("Training Logistic Regression model...")
model_lr.fit(X_train_resampled, y_train_resampled)
print("Model trained successfully.")

# 3. Lakukan prediksi pada set data pengujian
y_pred = model_lr.predict(X_test)

# 4. Hitung dan cetak metrik evaluasi
print("\n--- Model Evaluation --- ")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"Precision: {precision_score(y_test, y_pred):.4f}")
print(f"Recall: {recall_score(y_test, y_pred):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred):.4f}")

# 5. Hitung dan cetak confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

## Visualisasi Data

### Subtask:
Membuat visualisasi yang relevan untuk menjelaskan karakteristik data dan performa model. Ini dapat mencakup distribusi fitur, korelasi antar fitur, serta visualisasi hasil evaluasi model seperti heatmap confusion matrix atau kurva ROC.


```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Set a style for better aesthetics
sns.set_style("whitegrid")

# 1. Visualize initial class distribution of the target variable 'y'
plt.figure(figsize=(8, 6))
sns.countplot(x=y)
plt.title('Initial Class Distribution of Risk_Flag (Before SMOTE)')
plt.xlabel('Risk_Flag')
plt.ylabel('Count')
plt.show()

# 2. Visualize class distribution of 'y_train_resampled' after SMOTE
plt.figure(figsize=(8, 6))
sns.countplot(x=y_train_resampled)
plt.title('Class Distribution of Risk_Flag (After SMOTE)')
plt.xlabel('Risk_Flag')
plt.ylabel('Count')
plt.show()

# 3. Generate a heatmap for the confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix Heatmap')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

**Reasoning**:
To further evaluate the model's performance as per the subtask, I will now generate the ROC curve and calculate the Area Under the Curve (AUC). This will provide insight into the model's ability to distinguish between classes.



```python
from sklearn.metrics import roc_curve, roc_auc_score

# Calculate predicted probabilities for the positive class
y_pred_proba = model_lr.predict_proba(X_test)[:, 1]


fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)


auc = roc_auc_score(y_test, y_pred_proba)


plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

print(f"AUC for Logistic Regression model: {auc:.4f}")
```

## Penyimpanan dan Pengujian Model

### Subtask:
Menyimpan model Machine Learning yang telah dilatih. Kemudian, memuat kembali model yang telah disimpan dan mengujinya dengan data baru atau bagian dari data test yang belum pernah digunakan dalam proses pelatihan atau evaluasi.


```python
import joblib
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix


model_filename = 'logistic_regression_model.joblib'
joblib.dump(model_lr, model_filename)
print(f"Model saved successfully to {model_filename}")


loaded_model = joblib.load(model_filename)
print(f"Model loaded successfully from {model_filename}")


y_pred_loaded = loaded_model.predict(X_test)


print("\n--- Evaluation of Loaded Model ---")
print(f"Accuracy (Loaded Model): {accuracy_score(y_test, y_pred_loaded):.4f}")
print(f"Precision (Loaded Model): {precision_score(y_test, y_pred_loaded):.4f}")
print(f"Recall (Loaded Model): {recall_score(y_test, y_pred_loaded):.4f}")
print(f"F1-Score (Loaded Model): {f1_score(y_test, y_pred_loaded):.4f}")

print("\nConfusion Matrix (Loaded Model):")
print(confusion_matrix(y_test, y_pred_loaded))
```
